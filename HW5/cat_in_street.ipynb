```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

```python
street=mpimg.imread('street.png') 
cat = mpimg.imread('cat.png') 
```

```python
# write your code in this cell.
(street_height, street_width, _) = street.shape
(cat_height, cat_width, _) = cat.shape
start_x = 0
start_y = (street_height - cat_height)
for street_y in range(start_y, street_height) :
    for street_x in range(start_x, cat_width) :
        cat_x = street_x - start_x
        cat_y = street_y - start_y
        color = cat [cat_y][cat_x]
        if not sum([np.abs(color[0] - 1.0) < 0.01 for i in range(3)]):
            street[street_y-25][street_x] = color
```
