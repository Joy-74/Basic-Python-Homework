## Analyze the word frequency of online text

You need to remove the stop words, punctuation marks, capital letters.

```python
import urllib
url = "http://www.gutenberg.org/files/863/863-0.txt"
response = urllib.request.urlopen(url)
raw = response.read().decode('utf8')
i0 = raw.rfind("CHAPTER I. I GO TO STYLES")
i1 = raw.rfind("THE END")
raw = raw[i0:i1+7] # raw string
```

```python
len(raw)
```

```python
# write your code here, you may want to import new things
# plot the frequency in the end, as illustrated in HW7.pdf.

from nltk import word_tokenize, book
from nltk.corpus import stopwords
tokens = word_tokenize(raw)

stopwds = stopwords.words("english")
filtered_text = [w.lower() for w in tokens if w.lower() not in stopwds and w.isalnum()] 
filtered_text = nltk.Text(filtered_text)
fdist_filtered=book.FreqDist(filtered_text)
fdist_filtered.plot(30, cumulative=False)
```

## Create a Complete Directed Graph

```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
# implement the function create_complete_graph here
def create_complete_graph(n):
  G = nx.DiGraph()
  G.add_nodes_from([i for i in range(n)])
  for i in range(n):
    for j in range(n):
      G.add_edge(i,j)
  return G
```

```python
# test example
G = create_complete_graph(4)
nx.draw_networkx(G,with_labels=True)
plt.show()
```
