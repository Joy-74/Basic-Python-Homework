## Problem 1: Analyze the word frequency of online text

You need to remove the stop words, punctuation marks, capital letters.

```python
import urllib
url = "http://www.gutenberg.org/files/863/863-0.txt"
response = urllib.request.urlopen(url)
raw = response.read().decode('utf8')
i0 = raw.rfind("CHAPTER I. I GO TO STYLES")
i1 = raw.rfind("THE END")
raw = raw[i0:i1+7] # raw string
```

```python
from nltk.corpus import stopwords
from nltk import word_tokenize
words = word_tokenize(raw)
stop_words = stopwords.words('english')
filtered_words = [w.lower() for w in words if w.lower() not in stop_words and w.isalnum()]

filtered_words = nltk.Text(filtered_words)

word_freq = nltk.FreqDist(filtered_words)
word_freq.plot(30, cumulative=False)
```

```python
# write your code here, you may want to import new things
# plot the frequency in the end, as illustrated in HW7.pdf.
from nltk.corpus import stopwords
import matplotlib.pyplot as plt
from nltk import word_tokenize, book

# download NLTK stop words
stop_words = stopwords.words('english')

# split into words
words = word_tokenize(raw)

# Remove stop words, punctuation marks, and capital letters.
filtered_words = [w.lower() for w in words if w.lower() not in stop_words and w.isalnum()]
filtered_words = nltk.Text(filtered_words)

# Calculate word frequency
word_freq = nltk.FreqDist(filtered_words)

# Plot the frequency distribution
plt.xlabel('Samples')
plt.ylabel('Counts')
word_freq.plot(30, cumulative=False)


```

```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
# implement the function create_complete_graph here

def create_complete_graph(n):
    G = nx.DiGraph()
    nodes = range(n)
    G.add_nodes_from(nodes)
    for i in nodes:
        for j in nodes:
            G.add_edge(i, j)
    return G
```

```python
# test example in HW7.pdf
G = create_complete_graph(4)
nx.draw_networkx(G,with_labels=True)
plt.show()
```
