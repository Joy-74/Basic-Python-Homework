stopwords 停用词：停用词是搜索引擎被编程为忽略的常用词（例如“the”、“a”、“an”、“in”），无论是在索引条目进行搜索时还是在检索条目时作为搜索查询的结果。 
我们不希望这些词占用我们数据库中的空间，或占用宝贵的处理时间。为此，我们可以通过存储您认为是停用词的单词列表来轻松删除它们

```python
import nltk
nltk.download("stopwords")
```

```python
from nltk.corpus import stopwords
print(stopwords.words("english"))  # english: 可以替换别的语言
print(type(stopwords.words("english")))
```

But the punctuation marks are missing from the stop words

- `s.startswith(t)`	test if s starts with t 测试 s 是否以 t 开头
- `s.endswith(t)`	test if s ends with t 测试 s 是否以 t 结尾
- `t in s`	test if t is a substring of s 测试 t 是否是 s 的子串
- `s.islower()`	test if s contains cased characters and all are lowercase 测试 s 是否包含大小写字符且全部为小写
- `s.isupper()`	test if s contains cased characters and all are uppercase 测试 s 是否包含大小写字符且全部为大写
- `s.isalpha()`	test if s is non-empty and all characters in s are alphabetic 测试 s 是否非空并且 s 中的所有字符都是字母
- `s.isalnum()`	test if s is non-empty and all characters in s are alphanumeric 测试 s 是否为非空以及 s 中的所有字符是否为字母数字
- `s.isdigit()`	test if s is non-empty and all characters in s are digits 测试 s 是否非空并且 s 中的所有字符都是数字
- `s.istitle()`	test if s contains cased characters and is titlecased (i.e. all words in s have initial capitals) 测试 s 是否包含大小写字符并且是标题化的（即 s 中的所有单词都有首字母大写）

What about words with capital letters?

Let's convert all letters to lower-case.

The Python String lower() method converts all the case-based characters in a string into lowercased characters. 

```python
print("Pic16A".lower())
```

```python
stopwds = stopwords.words("english")
filtered_text1 = [w.lower() for w in book.text1 if w.lower() not in stopwds and w.isalnum()]  # filtered_text1 is a list of tokens
filtered_text1 = nltk.Text(filtered_text1)
fdist_filtered=book.FreqDist(filtered_text1)
fdist_filtered.plot(30, cumulative=False)
```
